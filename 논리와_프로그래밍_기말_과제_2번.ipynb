```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import matplotlib.pyplot as plt


Caffeine = "https://raw.githubusercontent.com/jeonjunhyeok/Jeonjunhyeok/main/caffeine.csv"
df = pd.read_csv(Caffeine) 


X = df.iloc[:,1:4]
Y = df.iloc[:,4]
Y = pd.get_dummies(Y)
Z= ((X-X.mean())/X.std())


model = Sequential()
model.add(Dense(1200,  input_dim=3, activation='relu'))
model.add(Dense(1050,  activation='relu'))
model.add(Dense(6, activation='softmax'))
model.summary()
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
history=model.fit(Z, Y, epochs=2000, batch_size=5)


plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],'b-',label='loss')
plt.xlabel('Epoch')
plt.legend()


plt.subplot(1,2,2)
plt.plot(history.history['accuracy'],'g-',label='accuracy')
plt.xlabel('Epoch')
plt.ylim(0.5,1)
plt.legend()


plt.show()
```
